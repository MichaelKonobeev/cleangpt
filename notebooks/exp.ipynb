```python
import os
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
from cleangpt.gpt import make as make_gpt
from cleangpt.exp.data import make_loader, most_recent_file, tokenize, DATA_PATH
from cleangpt.exp.train import Trainer
from u4ml.plot import LinesPlotter
```

```python
OUTPUT_PATH = os.path.join(DATA_PATH, "generated")
```

```python
model = make_gpt(vocab_size=129, seqlen=128, config_name="micro")
trainer = Trainer(model)
```

```python
dloader = make_loader(batch_size=64)
filename = most_recent_file()
```

```python
nepochs = 100
```

```python
def generate(model, inputs=dloader.dataset[0][0][None],
             num_new_tokens=len(dloader.dataset.tokens) - len(dloader.dataset[0][0]),
             output_path=os.path.join(OUTPUT_PATH, dloader.dataset.filename),
             **kwargs):
  model.eval()
  outputs = model.generate(inputs, num_new_tokens, **kwargs)
  text = tokenize(outputs[0].tolist(), inverse=True)
  if output_path is not None:
    with open(output_path, 'w', encoding="utf8") as outfile:
      outfile.write(text)
  print(text)
  model.train()
```

```python
plt.semilogy()
with LinesPlotter.make_autoclear_context() as plotter:
  with tqdm(total=nepochs) as pbar:
    for n in range(nepochs):
      for i, (inputs, targets, loss) in enumerate(trainer.iter_epoch(dloader)):
        if i % 36 == 0 or i in (0, len(dloader) - 1):
          plotter.extend("loss", [n * len(dloader) + i], [float(loss)])
      clear_output(True)
      pbar.update(1)
      display(pbar.container)
      display(plotter.output)
      generate(model, sample=True, topk=10)
```
